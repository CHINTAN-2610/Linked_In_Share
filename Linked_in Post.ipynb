#Following is simply prediction of GPA based on SAT and Rndom 1 2 3 variable

```python
#Step 1 is simply importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
#import LinearRegression from sklearn
from sklearn.linear_model import LinearRegression
```

```python
#impot StanderdScaller from sklearn
from sklearn.preprocessing import StandardScaler
```

```python
#Importing data from system stored in .csv file
raw_data = pd.read_csv("C:/Users/Chi_12345/The Data Science Course 2021 - All Resources/05/S34_L221/1.02. Multiple linear regression.csv")
```

```python
raw_data.head(10)
```

#Define inputs and targets for model

```python
data = raw_data.copy()
targets = data["GPA"]
inputs = data[["SAT","Rand 1,2,3"]]
```

```python
targets
```

```python
inputs
```

```python
#Followind methods are used for standerdization
scaler = StandardScaler()
scaler.fit(inputs)
scaler_inputs = scaler.transform(inputs)
scaler_inputs
```

#Lets fit LinearRegression model 

```python
reg = LinearRegression()
reg.fit(scaler_inputs , targets)
```

```python
#lets Create Summary Table for regression model
summary_table = pd.DataFrame(data = [["Bias"],["SAT"],["Rand 1,2,3"]] , columns = ['Features'])
summary_table["Weights"] = reg.intercept_ , reg.coef_[0] , reg.coef_[1]
summary_table
```

```python
#Now lets find p-values and F-values for features contribution to predict GPA
from sklearn.feature_selection import f_regression
p_values = f_regression(inputs , targets)[1]
f_values = f_regression(inputs , targets)[0]
```

```python
Table = pd.DataFrame(data = [["SAT"],["Rand 1,2,3"]] , columns = ['Features'])
Table["p_value"] = p_values
Table["f_value"] = f_values
Table
```
